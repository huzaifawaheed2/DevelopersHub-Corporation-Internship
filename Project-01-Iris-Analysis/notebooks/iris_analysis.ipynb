# Dataset Understanding and Description

## About the Dataset

The Iris Dataset is a multivariate dataset introduced by the British statistician and biologist Ronald Fisher in 1936. It is one of the most frequently used datasets for demonstrating statistical analysis, data visualization, and machine learning concepts.

The dataset contains measurements collected from iris flowers belonging to three different species:

1. Iris Setosa
2. Iris Versicolor
3. Iris Virginica

Each observation in the dataset represents a single flower sample and includes measurements of its sepal and petal dimensions.


## Dataset Features

The dataset contains six columns:

### 1. Id
A unique identifier assigned to each flower sample.

### 2. SepalLengthCm
The length of the sepal measured in centimeters.

### 3. SepalWidthCm
The width of the sepal measured in centimeters.

### 4. PetalLengthCm
The length of the petal measured in centimeters.

### 5. PetalWidthCm
The width of the petal measured in centimeters.

### 6. Species
The species classification of the iris flower:
- Iris-setosa
- Iris-versicolor
- Iris-virginica

## Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is the process of examining and visualizing data to discover patterns, relationships, trends, and potential anomalies. EDA helps transform raw data into meaningful insights and provides a deeper understanding of the dataset.

In this section, various visualization techniques will be used to analyze the characteristics of the Iris Dataset. Scatter plots, histograms, and box plots will help reveal relationships between features, understand data distributions, and identify variations within the dataset.

### Scatter Plot Analysis

Scatter plots are widely used to examine relationships between numerical variables and identify potential patterns within a dataset. By visualizing two features simultaneously, scatter plots help reveal trends, correlations, clusters, and relationships among observations.

In this analysis, scatter plots are used to explore the relationships between both petal measurements and sepal measurements across different iris flower species. These visualizations help identify similarities, differences, and grouping patterns within the dataset while providing a better understanding of species characteristics.

#### Petal Measurements Analysis

Petal measurements are among the most important features in the Iris Dataset because they often provide clear distinctions between flower species.

The following scatter plot examines the relationship between petal length and petal width across different iris flower species.

```python
# Create a scatter plot to analyze the relationship between petal length and petal width

plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=df,
    x="PetalLengthCm",
    y="PetalWidthCm",
    hue="Species"
)

plt.title("Relationship Between Petal Length and Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")

plt.show()
```

##### Observation

The scatter plot demonstrates a strong positive relationship between petal length and petal width. As petal length increases, petal width generally increases as well, indicating a positive correlation between these two features.

The visualization clearly separates the three iris flower species into distinct clusters. Iris-setosa samples are concentrated in the lower-left region of the plot, indicating the smallest petal dimensions. Iris-versicolor occupies the middle range, while Iris-virginica appears in the upper-right region with the largest petal measurements.

Although there is slight overlap between Iris-versicolor and Iris-virginica, the species remain largely distinguishable based on petal measurements. This suggests that petal length and petal width are highly informative features for differentiating iris species and may be valuable predictors in classification models.

#### Sepal Measurements Analysis

Sepal measurements also provide useful information about the physical characteristics of iris flowers. 

The following scatter plot examines the relationship between sepal length and sepal width to identify patterns and variations among different species.

```python
# Create a scatter plot to analyze the relationship between sepal length and sepal width

plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=df,
    x="SepalLengthCm",
    y="SepalWidthCm",
    hue="Species"
)

plt.title("Relationship Between Sepal Length and Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

plt.show()
```

##### Observation

The scatter plot illustrates the relationship between sepal length and sepal width across the three iris flower species. Compared to petal measurements, the species are less distinctly separated, resulting in a greater degree of overlap among the observations.

Iris-setosa forms a relatively distinct cluster with higher sepal widths and shorter sepal lengths. In contrast, Iris-versicolor and Iris-virginica exhibit considerable overlap, making them more difficult to distinguish based solely on sepal measurements.

The visualization suggests that sepal length and sepal width provide useful information about flower characteristics; however, these features are less effective for species differentiation than petal measurements. This observation highlights the greater discriminative power of petal-related features within the Iris Dataset.

### Histogram Analysis

Histograms are useful for understanding the distribution of numerical variables within a dataset. They show how frequently different values occur and help identify patterns such as concentration, spread, skewness, and variability in the data.

In this analysis, histograms are used to examine the distributions of both sepal length and petal length measurements. Understanding these distributions provides valuable insight into the characteristics of the iris flower samples and helps reveal how the measurements are distributed across the dataset.

#### Sepal Length Distribution Analysis

Sepal length is one of the primary measurements recorded in the Iris Dataset. Examining its distribution helps identify the range of values, common measurement intervals, and the overall spread of observations.

The following histogram visualizes the distribution of sepal length measurements across all iris flower samples.

```python
# Create a histogram to examine the distribution of sepal length measurements

plt.figure(figsize=(8, 6))

sns.histplot(
    data=df,
    x="SepalLengthCm",
    bins=10,
    kde=True
)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")

plt.show()
```

##### Observation

The histogram shows the distribution of sepal length measurements across all iris flower samples. Most observations are concentrated between approximately 5.0 cm and 6.5 cm, indicating that this range represents the most common sepal lengths within the dataset.

The distribution appears relatively balanced with a slight concentration around the central values. Fewer observations are present at the lower and higher ends of the range, suggesting that extremely small or extremely large sepal lengths are less common.

The KDE curve further highlights the overall shape of the distribution and indicates that the dataset contains a reasonable spread of sepal length measurements. This variability is beneficial for exploratory analysis because it reflects differences among the iris flower species.

#### Petal Length Distribution Analysis

Petal length is one of the most important features in the Iris Dataset and plays a significant role in distinguishing between different flower species. Analyzing its distribution helps identify common value ranges and understand the variability of petal measurements.

The following histogram visualizes the distribution of petal length measurements across all iris flower samples.

```python
# Create a histogram to examine the distribution of petal length measurements

plt.figure(figsize=(8, 6))

sns.histplot(
    data=df,
    x="PetalLengthCm",
    bins=10,
    kde=True
)

plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")

plt.show()
```

##### Observation

The histogram illustrates the distribution of petal length measurements across all iris flower samples. Unlike the sepal length distribution, the petal length distribution appears to contain multiple peaks, indicating the presence of distinct groups within the dataset.

A large concentration of observations is visible between approximately 1.0 cm and 2.0 cm, while another major concentration occurs between approximately 4.0 cm and 6.0 cm. This pattern suggests that different iris species possess significantly different petal length characteristics.

The KDE curve further emphasizes the separation between these groups, highlighting the strong variation in petal length measurements. This observation supports the findings from the scatter plots, where petal-related features demonstrated strong discriminative power for distinguishing between iris flower species.

Overall, petal length appears to be one of the most informative features in the dataset and is highly useful for species differentiation and classification tasks.

### Box Plot Analysis

Box plots are useful for examining the distribution, spread, and variability of numerical features. They provide a visual summary of the data by displaying the median, quartiles, overall range, and potential outliers.

In this analysis, box plots are used to compare sepal length and petal length measurements across different iris flower species. These visualizations help identify differences in feature distributions and highlight species-specific characteristics.

#### Sepal Length Distribution by Species

Sepal length is an important measurement that varies among iris flower species. Comparing the distribution of sepal length across species helps identify differences in central tendency, spread, and potential outliers.

The following box plot visualizes the distribution of sepal length measurements for each iris flower species.

```python
# Create a box plot to compare sepal length distributions across species

plt.figure(figsize=(8, 6))

sns.boxplot(
    data=df,
    x="Species",
    y="SepalLengthCm"
)

plt.title("Sepal Length Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")

plt.show()
```

##### Observation

The box plot highlights differences in sepal length distributions among the three iris flower species. Iris-setosa generally exhibits the smallest sepal lengths, while Iris-virginica displays the largest measurements. Iris-versicolor occupies an intermediate range between the other two species.

The median sepal length increases progressively from Iris-setosa to Iris-virginica, indicating distinct species characteristics. Additionally, the spread of values differs across species, reflecting natural variation within each group.

A potential outlier is visible within the Iris-virginica category, suggesting the presence of an observation with a sepal length notably different from the majority of samples in that species. Overall, the visualization demonstrates that sepal length varies across species and contributes useful information for species differentiation.

#### Petal Length Distribution by Species

Petal length is one of the most distinguishing characteristics of iris flowers. Comparing petal length distributions across species provides valuable insight into species-specific differences and helps evaluate how effectively this feature separates the flower categories.

The following box plot visualizes the distribution of petal length measurements for each iris flower species.

```python
# Create a box plot to compare petal length distributions across species

plt.figure(figsize=(8, 6))

sns.boxplot(
    data=df,
    x="Species",
    y="PetalLengthCm"
)

plt.title("Petal Length Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")

plt.show()
```

##### Observation

The box plot reveals clear differences in petal length distributions among the three iris flower species. Iris-setosa exhibits the smallest petal lengths, Iris-versicolor occupies the intermediate range, and Iris-virginica shows the largest petal length measurements.

The median petal length increases substantially from Iris-setosa to Iris-virginica, demonstrating strong variation between species. Unlike sepal length, the petal length distributions show minimal overlap, making this feature highly effective for distinguishing between iris flower categories.

A few potential outliers are visible within the Iris-setosa and Iris-versicolor groups. However, these observations do not significantly affect the overall distribution patterns. The clear separation between species indicates that petal length is one of the most informative features in the dataset and can play an important role in classification and predictive modeling tasks.
