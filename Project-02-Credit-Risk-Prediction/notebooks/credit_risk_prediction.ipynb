# Credit Risk Prediction

## Introduction

Credit risk prediction is an important application of data science and machine learning in the banking and financial sector.

Banks and financial institutions receive many loan applications from people with different financial backgrounds. Before approving a loan, it is important to evaluate whether an applicant is likely to repay the loan or may face difficulties in repayment.

Making incorrect loan approval decisions can lead to financial losses for the organization. Therefore, financial institutions use applicant information such as income, education level, employment status, loan amount, and credit history to support their decision-making process.

With the help of data analysis and machine learning, it is possible to identify patterns in historical loan data and build models that can predict loan approval outcomes more efficiently and consistently.

In this project, the Loan Prediction Dataset will be analyzed to understand applicant characteristics, perform data cleaning, conduct exploratory data analysis (EDA), and develop classification models for predicting loan approval status.

## Problem Statement

Loan approval is one of the most important decisions for banks and financial institutions. Approving a loan for a high-risk applicant may result in financial loss, while rejecting a genuine applicant may lead to missed business opportunities.

The objective of this project is to analyze applicant information and identify the factors that influence loan approval decisions.

Using historical loan application data, machine learning models will be developed to predict whether a loan application is likely to be approved or not approved.

The final model will be evaluated using classification performance metrics to measure its predictive capability and effectiveness.

### Identify Mode Values for Categorical Features

Before handling missing values in categorical features, the most frequent value (mode) is identified.

These values will be used to replace missing entries and maintain the overall distribution of the dataset.

```python
# Display mode values for categorical features

print("Gender Mode:", df["Gender"].mode()[0])

print("Married Mode:", df["Married"].mode()[0])

print("Dependents Mode:", df["Dependents"].mode()[0])

print("Self_Employed Mode:", df["Self_Employed"].mode()[0])
```

#### Observation

The mode values were identified for all categorical features containing missing data.

The most frequent categories are Male for Gender, Yes for Married, 0 for Dependents, and No for Self_Employed.

These values represent the most common categories in their respective features and will be used to replace missing entries during the data cleaning process.

### Identify Median Values for Numerical Features

Before handling missing values in numerical features, the median value of each column is calculated.

The median is preferred because it is resistant to extreme values and provides a reliable measure of central tendency.

```python
# Display median values for numerical features

print("LoanAmount Median:", df["LoanAmount"].median())

print("Loan Amount Term Median:", df["Loan_Amount_Term"].median())

print("Credit History Median:", df["Credit_History"].median())
```

#### Observation

The median values were calculated for all numerical features containing missing data.

The median loan amount is 128.0, the median loan amount term is 360.0 months, and the median credit history value is 1.0.

These median values will be used to replace missing records because they are less sensitive to extreme values and help preserve the overall distribution of the dataset.

### Handling Missing Values

After identifying the mode and median values, missing records can now be replaced using appropriate imputation techniques.

Categorical features will be filled using their mode values, while numerical features will be filled using their median values.

This process ensures that the dataset becomes complete and ready for further analysis and model development.

```python
# Fill missing values in categorical features using mode

df_clean["Gender"] = df_clean["Gender"].fillna(df_clean["Gender"].mode()[0])

df_clean["Married"] = df_clean["Married"].fillna(df_clean["Married"].mode()[0])

df_clean["Dependents"] = df_clean["Dependents"].fillna(df_clean["Dependents"].mode()[0])

df_clean["Self_Employed"] = df_clean["Self_Employed"].fillna(df_clean["Self_Employed"].mode()[0])
```

```python
# Fill missing values in numerical features using median

df_clean["LoanAmount"] = df_clean["LoanAmount"].fillna(df_clean["LoanAmount"].median())

df_clean["Loan_Amount_Term"] = df_clean["Loan_Amount_Term"].fillna(df_clean["Loan_Amount_Term"].median())

df_clean["Credit_History"] = df_clean["Credit_History"].fillna(df_clean["Credit_History"].median())
```

#### Observation

Missing values in categorical features were replaced using the most frequently occurring category (mode).

Missing values in numerical features were replaced using the median value of each feature.

The dataset should now contain no missing values and is ready for verification.

### Verify Missing Values

#### Observation

### Encoding Categorical Variables

Machine learning algorithms require numerical input and cannot directly process categorical text values.

Therefore, categorical features must be converted into numerical representations before model training.

Before encoding, the unique values of each categorical feature are examined to understand the categories present in the dataset.

```python
# Display unique values of categorical features before encoding

categorical_columns = [
    "Gender",
    "Married",
    "Dependents",
    "Education",
    "Self_Employed",
    "Property_Area",
    "Loan_Status"
]

for column in categorical_columns:
    print(f"{column}: {df_clean[column].unique().tolist()}")
```

#### Observation

The categorical features contain a limited number of distinct categories, making them suitable for encoding.

Most variables are binary in nature, such as Gender, Married, Self_Employed, and Loan_Status, which contain only two categories each.

The Dependents feature contains four categories (0, 1, 2, and 3+), while Property_Area contains three categories (Urban, Rural, and Semiurban).

Since machine learning algorithms require numerical input, these categorical values will be converted into numerical representations during the encoding stage.

```python
# Encode binary categorical features

df_clean["Gender"] = df_clean["Gender"].map({
    "Male": 1,
    "Female": 0
})

df_clean["Married"] = df_clean["Married"].map({
    "Yes": 1,
    "No": 0
})

df_clean["Education"] = df_clean["Education"].map({
    "Graduate": 1,
    "Not Graduate": 0
})

df_clean["Self_Employed"] = df_clean["Self_Employed"].map({
    "Yes": 1,
    "No": 0
})

df_clean["Loan_Status"] = df_clean["Loan_Status"].map({
    "Y": 1,
    "N": 0
})
```

```python
# Encode multi-category features

df_clean["Dependents"] = df_clean["Dependents"].map({
    "0": 0,
    "1": 1,
    "2": 2,
    "3+": 3
})

df_clean["Property_Area"] = df_clean["Property_Area"].map({
    "Rural": 0,
    "Semiurban": 1,
    "Urban": 2
})
```

## Model Training and Testing

After completing data cleaning and exploratory data analysis, the dataset is prepared for machine learning model development.

The objective is to build a classification model that can predict whether a loan application is likely to be approved or rejected based on applicant characteristics and financial information.

### Feature Selection

Before training the machine learning model, the dataset is divided into input features (X) and the target variable (y).

The target variable is Loan_Status, while all remaining variables are used as predictor features for loan approval prediction.

```python
# Define input features and target variable

X = df_clean.drop("Loan_Status", axis=1)

y = df_clean["Loan_Status"]
```

#### Verify Feature Matrix and Target Variable

The dimensions of the feature matrix and target variable are checked to ensure that the dataset has been prepared correctly for model training.

```python
# Display dataset dimensions

print("Feature Matrix Shape:", X.shape)

print("Target Variable Shape:", y.shape)
```

##### Observation

The feature matrix contains 614 observations and 11 predictor variables that will be used for model training.

The target variable contains 614 corresponding loan approval records.

This confirms that the dataset has been correctly separated into input features and the target variable and is ready for train-test splitting.

### Train-Test Split

Before training the machine learning model, the dataset is divided into training and testing subsets.

The training dataset is used to teach the model patterns and relationships within the data, while the testing dataset is used to evaluate model performance on unseen data.

An 80:20 split ratio is used, where 80% of the data is allocated for training and 20% for testing.

A fixed random state of 42 is applied to ensure reproducibility of results.

```python
# Split the dataset into training and testing subsets
# 80% of the data is used for training and 20% for testing
# random_state ensures consistent results across runs

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

#### Verify Train-Test Split

The dimensions of the training and testing datasets are examined to confirm that the split was performed successfully.

Verifying the dataset shapes ensures that the data is ready for machine learning model training and evaluation.

```python
# Display train-test split dimensions

print("X_train Shape:", X_train.shape)
print("X_test Shape :", X_test.shape)

print("y_train Shape:", y_train.shape)
print("y_test Shape :", y_test.shape)
```

##### Observation

The dataset was successfully divided into training and testing subsets using an 80:20 ratio.

The training dataset contains 491 observations and 11 predictor variables, while the testing dataset contains 123 observations and 11 predictor variables.

Similarly, the target variable was divided into 491 training records and 123 testing records.

The dataset is now ready for machine learning model training and performance evaluation.

### Logistic Regression Model

Logistic Regression is a supervised machine learning algorithm commonly used for binary classification problems.

Since the target variable (Loan_Status) contains two classes representing approved and rejected loan applications, Logistic Regression is an appropriate model for predicting loan approval outcomes.

The model will be trained using the training dataset and evaluated using the testing dataset.

#### Model Training

The Logistic Regression model is created and trained using the training dataset.

During the training process, the model learns patterns and relationships between the predictor variables and the target variable (Loan_Status).

These learned patterns are later used to make predictions on unseen data.

```python
# Create Logistic Regression model

logistic_model = LogisticRegression(max_iter=10000)

# Train the model

logistic_model.fit(X_train, y_train)
```

##### Observation

The model successfully generated predictions for the testing dataset.

These predicted values will be compared with the actual loan approval outcomes to evaluate the model's performance.

#### Model Prediction

After training the Logistic Regression model, predictions are generated using the testing dataset.

These predictions represent the model's estimated loan approval outcomes and will be compared with the actual values during model evaluation.

```python
# Generate predictions on testing data

y_pred = logistic_model.predict(X_test)

y_pred[:20]
```

##### Observation

The Logistic Regression model successfully generated predictions for the testing dataset.

The predicted values represent the model's estimated loan approval outcomes and include both approved and rejected loan applications.

These predictions will be compared with the actual values to evaluate model performance.

#### Accuracy Score

Accuracy measures the proportion of correctly classified observations among all predictions.

It provides an overall indication of how well the model performs on the testing dataset.

```python
# Calculate model accuracy

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy Score:", round(accuracy * 100, 2), "%")
```

##### Observation

The Logistic Regression model achieved an accuracy score of 78.86% on the testing dataset.

This indicates that approximately 79 out of every 100 loan applications were classified correctly.

The result suggests that the model has learned meaningful patterns from the dataset and demonstrates satisfactory predictive performance.

### Confusion Matrix

A confusion matrix provides a detailed breakdown of the model's predictions by comparing the predicted values with the actual values.

It helps identify correctly classified observations as well as classification errors.

```python
# Generate confusion matrix

cm = confusion_matrix(y_test, y_pred)

# Create readable confusion matrix table

cm_df = pd.DataFrame(
    cm,
    index=["Actual Rejected", "Actual Approved"],
    columns=["Predicted Rejected", "Predicted Approved"]
)

cm_df
```

#### Observation

The confusion matrix shows that the Logistic Regression model correctly classified 18 rejected loan applications and 79 approved loan applications.

The model incorrectly classified 25 rejected applications as approved and only 1 approved application as rejected.

Overall, the model performs well in identifying approved loan applications, although some rejected applications are misclassified as approved.

A confusion matrix heatmap is created to provide a visual representation of the model's classification performance.

### Classification Report

The classification report provides additional evaluation metrics, including precision, recall, and F1-score.

These metrics offer a more detailed assessment of the model's classification performance for each class.
