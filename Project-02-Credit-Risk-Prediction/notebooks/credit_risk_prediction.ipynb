# Credit Risk Prediction

## Introduction

Credit risk prediction is an important application of data science and machine learning in the banking and financial sector.

Banks and financial institutions receive many loan applications from people with different financial backgrounds. Before approving a loan, it is important to evaluate whether an applicant is likely to repay the loan or may face difficulties in repayment.

Making incorrect loan approval decisions can lead to financial losses for the organization. Therefore, financial institutions use applicant information such as income, education level, employment status, loan amount, and credit history to support their decision-making process.

With the help of data analysis and machine learning, it is possible to identify patterns in historical loan data and build models that can predict loan approval outcomes more efficiently and consistently.

In this project, the Loan Prediction Dataset will be analyzed to understand applicant characteristics, perform data cleaning, conduct exploratory data analysis (EDA), and develop classification models for predicting loan approval status.

## Problem Statement

Loan approval is one of the most important decisions for banks and financial institutions. Approving a loan for a high-risk applicant may result in financial loss, while rejecting a genuine applicant may lead to missed business opportunities.

The objective of this project is to analyze applicant information and identify the factors that influence loan approval decisions.

Using historical loan application data, machine learning models will be developed to predict whether a loan application is likely to be approved or not approved.

The final model will be evaluated using classification performance metrics to measure its predictive capability and effectiveness.

### Missing Values Analysis

Missing values can affect the quality and performance of machine learning models.

Before handling missing values, it is important to identify which features contain missing data and how many values are missing in each feature.

```python
# Display the number of missing values in each column

missing_values = df.isnull().sum()

missing_values
```

#### Observation

The missing values analysis shows that several columns contain incomplete data.

The Credit_History column has the highest number of missing values (50), followed by Self_Employed (32), LoanAmount (22), Dependents (15), Loan_Amount_Term (14), Gender (13), and Married (3).

The remaining columns do not contain any missing values.

Since machine learning models cannot handle missing values directly, these missing records will need to be treated during the data cleaning phase.

### Missing Values Percentage

In addition to counting missing values, it is useful to calculate the percentage of missing data in each feature.

This helps determine the severity of missing values and supports appropriate data cleaning decisions.

```python
# Calculate and display the percentage of missing values in each column

missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_percentage.sort_values(ascending=False)
```

#### Observation

The percentage of missing values is relatively low across all features.

The Credit_History column contains the highest percentage of missing values (8.14%), followed by Self_Employed (5.21%) and LoanAmount (3.58%).

The remaining columns contain less than 3% missing data, while several features have no missing values at all.

Since the percentage of missing values is relatively small, appropriate imputation techniques can be applied during the data cleaning phase without significantly affecting the dataset quality.

### Duplicate Records Analysis

Duplicate records can reduce data quality and may lead to biased analytical results and machine learning predictions.

Before proceeding with data cleaning and model development, it is important to verify whether duplicate records exist in the dataset.

```python
# Calculate the total number of duplicate records in the dataset
duplicate_records = df.duplicated().sum()

print("Duplicated Records: ", duplicate_records)
```

#### Observation

The duplicate records analysis shows that there are no duplicate entries in the dataset.

Each record represents a unique loan application, ensuring that the dataset maintains high data quality and reducing the risk of biased analysis or model predictions.

Since no duplicate records were found, no duplicate removal is required during the data cleaning process.

```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

### Missing Values Visualization

Visualizing missing values provides a clear overview of incomplete data within the dataset.

This analysis helps identify features that require data cleaning and supports informed decisions regarding missing value treatment.

```python
# Calculate missing values for each feature
missing_values = df.isnull().sum().sort_values(ascending=False)

# Create a bar chart to visualize missing values
plt.figure(figsize=(6, 4))

plt.bar(
    missing_values.index,      # Feature names (X-axis)
    missing_values.values      # Missing values count (Y-axis)
)

plt.title("Missing Values by Feature")
plt.xlabel("Features")
plt.ylabel("Number of Missing Values")

plt.xticks(rotation=90)

plt.show()
```

#### Observation

The visualization shows that Credit_History contains the highest number of missing values in the dataset, followed by Self_Employed and LoanAmount.

Several features such as Education, Loan_ID, ApplicantIncome, CoapplicantIncome, Property_Area, and Loan_Status do not contain any missing values.

Overall, the dataset contains a relatively small amount of missing data, which can be handled effectively during the data preprocessing stage.
