# Credit Risk Prediction

## Introduction

Credit risk prediction is an important application of data science and machine learning in the banking and financial sector.

Banks and financial institutions receive many loan applications from people with different financial backgrounds. Before approving a loan, it is important to evaluate whether an applicant is likely to repay the loan or may face difficulties in repayment.

Making incorrect loan approval decisions can lead to financial losses for the organization. Therefore, financial institutions use applicant information such as income, education level, employment status, loan amount, and credit history to support their decision-making process.

With the help of data analysis and machine learning, it is possible to identify patterns in historical loan data and build models that can predict loan approval outcomes more efficiently and consistently.

In this project, the Loan Prediction Dataset will be analyzed to understand applicant characteristics, perform data cleaning, conduct exploratory data analysis (EDA), and develop classification models for predicting loan approval status.

## Problem Statement

Loan approval is one of the most important decisions for banks and financial institutions. Approving a loan for a high-risk applicant may result in financial loss, while rejecting a genuine applicant may lead to missed business opportunities.

The objective of this project is to analyze applicant information and identify the factors that influence loan approval decisions.

Using historical loan application data, machine learning models will be developed to predict whether a loan application is likely to be approved or not approved.

The final model will be evaluated using classification performance metrics to measure its predictive capability and effectiveness.

### Identify Mode Values for Categorical Features

Before handling missing values in categorical features, the most frequent value (mode) is identified.

These values will be used to replace missing entries and maintain the overall distribution of the dataset.

```python
# Display mode values for categorical features

print("Gender Mode:", df["Gender"].mode()[0])

print("Married Mode:", df["Married"].mode()[0])

print("Dependents Mode:", df["Dependents"].mode()[0])

print("Self_Employed Mode:", df["Self_Employed"].mode()[0])
```

```python
# Display mode values for categorical features

print("Gender Mode:", df["Gender"].mode()[0])

print("Married Mode:", df["Married"].mode()[0])

print("Dependents Mode:", df["Dependents"].mode()[0])

print("Self_Employed Mode:", df["Self_Employed"].mode()[0])
```

#### Observation

The mode values were identified for all categorical features containing missing data.

The most frequent categories are Male for Gender, Yes for Married, 0 for Dependents, and No for Self_Employed.

These values represent the most common categories in their respective features and will be used to replace missing entries during the data cleaning process.

### Identify Median Values for Numerical Features

Before handling missing values in numerical features, the median value of each column is calculated.

The median is preferred because it is resistant to extreme values and provides a reliable measure of central tendency.

```python
# Display median values for numerical features

print("LoanAmount Median:", df["LoanAmount"].median())

print("Loan Amount Term Median:", df["Loan_Amount_Term"].median())

print("Credit History Median:", df["Credit_History"].median())
```

#### Observation

The median values were calculated for all numerical features containing missing data.

The median loan amount is 128.0, the median loan amount term is 360.0 months, and the median credit history value is 1.0.

These median values will be used to replace missing records because they are less sensitive to extreme values and help preserve the overall distribution of the dataset.

### Handling Missing Values

After identifying the mode and median values, missing records can now be replaced using appropriate imputation techniques.

Categorical features will be filled using their mode values, while numerical features will be filled using their median values.

This process ensures that the dataset becomes complete and ready for further analysis and model development.

```python
# Fill missing values in categorical features using mode

df_clean["Gender"] = df_clean["Gender"].fillna(df_clean["Gender"].mode()[0])

df_clean["Married"] = df_clean["Married"].fillna(df_clean["Married"].mode()[0])

df_clean["Dependents"] = df_clean["Dependents"].fillna(df_clean["Dependents"].mode()[0])

df_clean["Self_Employed"] = df_clean["Self_Employed"].fillna(df_clean["Self_Employed"].mode()[0])
```

```python
# Fill missing values in numerical features using median

df_clean["LoanAmount"] = df_clean["LoanAmount"].fillna(df_clean["LoanAmount"].median())

df_clean["Loan_Amount_Term"] = df_clean["Loan_Amount_Term"].fillna(df_clean["Loan_Amount_Term"].median())

df_clean["Credit_History"] = df_clean["Credit_History"].fillna(df_clean["Credit_History"].median())
```

#### Observation

Missing values in categorical features were replaced using the most frequently occurring category (mode).

Missing values in numerical features were replaced using the median value of each feature.

The dataset should now contain no missing values and is ready for verification.

### Verify Missing Values

After completing the imputation process, the dataset is checked again to confirm that all missing values have been successfully handled.

```python
# Verify that all missing values have been removed

df_clean.isnull().sum()
```

#### Observation

The verification results confirm that all missing values have been successfully handled.

After applying mode-based imputation for categorical features and median-based imputation for numerical features, no missing values remain in the dataset.

The dataset is now complete and ready for feature transformation, exploratory data analysis, and machine learning model development.
